# Scikit-learn exercises

```python
import seaborn as sns

mpg = sns.load_dataset("mpg")
print(mpg.info())
```

```python
print(mpg.describe())
```

```python
# Identify rows with null values in the 'horsepower' column
rows_with_null_horsepower = mpg[mpg['horsepower'].isnull()]

# Display the rows with null values in the 'horsepower' column
print(rows_with_null_horsepower)
```

Guessing what their horsepower is, isn't reliable and could hurt our dataset, because they are missing we will drop them from the dataset completely, they can always be added later if the data were to surface and the data integrity of the data would stay intact.

```python
mpg.dropna(subset=["horsepower"], inplace=True)

print(mpg["horsepower"].isnull().sum())
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
sns.boxplot(x="cylinders", y="horsepower", data=mpg)
plt.title("Horsepower vs. Cylinders")
plt.xlabel("Cylinders")
plt.ylabel("Horsepower")
plt.show()
```
