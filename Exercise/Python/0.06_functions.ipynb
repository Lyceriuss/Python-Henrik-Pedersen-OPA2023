## Mathematical functions (*)
Make the following functions with def or lambda and plot their graphs in the same figure window.   

a)   F(x) = x**2 -3 

b)   g(x) = 4z-7


```python
import numpy as np
import matplotlib.pyplot as plt

# X^2 -3 = 4x -7 
# x*2 -4x +4 = 0
# X - 2x + 2 =0 >>       X-2 = 0     >> X = 2
# The solution to the function is 2 thats their relationship.

f = lambda x1: x1**2 -3
g = lambda x2: 4*x2-7

x_values = np.linspace(-10, 10, 400)


y_F = f(x_values)
y_g = g(x_values)

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(x_values, y_F, label="F(x) = x**2 - 3")
plt.plot(x_values, y_g, label="g(x) = 4x - 7")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphs of F(x) and g(x)")
plt.legend()

plt.grid(True)
plt.show()
```

## Change (**)
Create a function that takes a value as input parameter and print out the banknotes and coins in Swedish currency representing this value. For example 5289 would give the following printout:


- 5st 1000-lapp
- 1st 200-lapp
- 1st 50-lapp
- 1st 20-lapp
- 1st 10-krona
- 1st 5-krona
- 2st 2-krona

Now let the user input a value, and use the function to calculate the change.

```python
def ATM(Sum):
    
    BankNotes = [
        ("1000-lapp"),("200-lapp"),("50-lapp"),("20-lapp"),("10-krona"),("5-krona"),("2-krona")]
    
    
    Deposit = Sum
    Results = []
    
    while Deposit >0:
        if Deposit >= 1000:
            Results.append(BankNotes[0])
            Deposit -= 1000
        elif Deposit >= 200:
            Results.append(BankNotes[1])
            Deposit -= 200
        elif Deposit >= 50:
            Results.append(BankNotes[2])
            Deposit -= 50
        elif Deposit >= 20:
            Results.append(BankNotes[3])
            Deposit -= 20
        elif Deposit >= 10:
            Results.append(BankNotes[4])
            Deposit -= 10
        elif Deposit >= 5:
            Results.append(BankNotes[5])
            Deposit -= 5
        elif Deposit > 2:
            Results.append(BankNotes[6])
            Deposit -= 2
        else:
            break
            
    return Results


Withdrawal = ATM(5289)

banknote_counts = {}
for note in Withdrawal:
    if note in banknote_counts:
        banknote_counts[note] += 1
    else:
        banknote_counts[note] = 1

for note, count in banknote_counts.items():
    print(f"{count}st {note}")
    
```
