```python
import numpy as np
#A
dice_faces = np.array([1, 2, 3, 4, 5, 6])

mean_value = np.mean(dice_faces)
mean_value


```

```python
import matplotlib.pyplot as plt
#B
DiceRolls = [10, 100, 1000, 10000, 100000, 1000000, 10000000]

means = []

for n in DiceRolls:
    rolls = np.random.randint(1, 7, size=n)

    means.append(np.mean(rolls))

plt.figure(figsize=(10, 6))
plt.plot(DiceRolls, means, marker="o", linestyle="-", color="blue")
plt.xscale("log") 
plt.xlabel("Number of dice rolled (log scale)")
plt.ylabel("Mean of dice rolls")
plt.title("Mean of Dice Rolls Versus Number of Dice")
plt.grid(True)
plt.show()
```
