## Dice rolls (*)
Create a textfile called dice_rolls.txt using Python. Also for each subtask, write adequate headers.

  a)   Simulate 20 dice rolls and write them to your textfile. (*)

  b)   Sort the dice rolls from a) and write them to a separate row in the same textfile. (*)

  c)   Count the number of fours in the dice rolls and write them to a separate row in the same textfile. (*)

```python
import random


Rolls = [random.randint(1, 6) for _ in range(20)]


file_name = "0.08_FileHandlingData.txt"

with open(file_name, "w") as file:

    file.write("Dice Rolls:\n")
    file.writelines(f"{roll}\n" for roll in Rolls)  #Lambda

    sorted_rolls = sorted(Rolls)
    file.write("\nSorted Rolls:\n")
    file.writelines(f"{roll}\n" for roll in sorted_rolls)

    fours = Rolls.count(4)
    file.write(f"\nNumber of Fours: {fours}\n")
```

## Test results (*)
Read in the file test_result.txt (located in the data folder of this repo) in Python.

a)   Print out the text in the terminal (*)

b)   Create new rows in the same file and write the people and their corresponding scores in alphabetical order. (*)

c)   Create additional rows in the same file and sort the people after their grades. The grade limits are: (**)

```python

file_name = "0.08_test_result.txt"


# So you can test it!
original_data = [
    ("Adam Gustafsson", 25),
    ("Emil Johansson", 23),
    ("Sven Erik Karlsson", 13),
    ("Ove Karlsten", 41),
    ("Emma Boden", 32),
    ("Ida Håkansson", 23),
    ("Ella Ester", 41),
    ("Hanna Karlsson", 23),
    ("Johan Johansson", 42),
    ("Sven Erik Lundin", 39),
    ("Björn Björnsson", 39),
    ("Karl Karlsson", 32),
    ("Bose Bosseson", 32),
    ("Håkan Håkanson", 24),
    ("Jonas Jonasson", 31),
    ("Erik Eriksson", 31),
    ("Gore Bord", 55),
    ("Jakob Kallander", 65)
]

with open(file_name, "w") as file:
    for name, age in original_data:
        file.write(f"{name} {age}\n")




def gradeSort(score):
    if score < 20:
        return "F"
    elif score < 30:
        return "E"
    elif score < 40:
        return "D"
    elif score < 50:
        return "C"
    elif score < 60:
        return "B"
    elif score < 70:
        return "A"
    else:
        print("Invalid")

#A)
with open(file_name, "r") as file:  # "r" for read
    lines = file.readlines()
    print(lines)

#B)
with open(file_name, "a") as file: # "a" for append
    ABC_list = sorted(lines,key=str.lower)
    file.write("\nAlphabetical order:\n")
    file.writelines(ABC_list)
    
  
#C)
lines_with_grades = []
for line in lines:
    if not line.strip():  # Skip empty lines
        continue
    name, score = line.rsplit(None, 1)
    grade = gradeSort(int(score))
    lines_with_grades.append(f"{name} {score} {grade}\n")
    

with open(file_name, "a") as file:
    file.write("\nAlphabetical order + Grade:\n")
    file.write("\n")
    file.writelines(lines_with_grades)


    
```

## National test (*)
Read in the file NPvt19Ma2A.txt and NPvt19Ma2C.txt (located in the data folder of this repo) in Python. Use matplotlib to plot pie charts for each grade categories in each file.

```python
import matplotlib.pyplot as plt

file_name1 = "0.08_NPvt19Ma2A.txt"
file_name2 = "0.08_NPvt19Ma2C.txt"

with open(file_name1, "r") as file:  # "r" for read
    Ma2A = file.readlines()
    
with open(file_name2, "r") as file:  # "r" for read
    Ma2C = file.readlines()
    
    
def parse_data_and_plot(data, title):
    grades = []
    percentages = []

    for line in data:
        parts = line.split()
        grade = parts[0]
        percentage = float(parts[1].strip("%"))

        if percentage != 0.0:       #Lets not have 0 points in oour chart
            grades.append(grade)
            percentages.append(percentage)
        
    plt.figure(figsize=(8, 8))
    plt.pie(percentages, labels=grades, autopct="%1.1f%%", startangle=90)
    plt.title(title)
    plt.show()


parse_data_and_plot(Ma2A, "Grade Distribution for Ma2A")
parse_data_and_plot(Ma2C, "Grade Distribution for Ma2C")
```
